
# Visibility of Advertisement for Customers

Using an company's advertising data set, try to predict whether or not a particular internet user clicked on an Advertisement based off the features of that user.

This data set contains the following features:

* 'Daily Time Spent on Site': consumer time on site in minutes
* 'Age': cutomer age in years
* 'Area Income': Avg. Income of geographical area of consumer
* 'Daily Internet Usage': Avg. minutes a day consumer is on the internet
* 'Ad Topic Line': Headline of the advertisement
* 'City': City of consumer
* 'Male': Whether or not consumer was male
* 'Country': Country of consumer
* 'Timestamp': Time at which consumer clicked on Ad or closed window
* 'Clicked on Ad': 0 or 1 indicated clicking on Ad

### Logistic Regression

**Split the data into training set and testing set using train_test_split**

```python
ad_data = pd.read_csv('advertising.csv')
#X = ad_data[['Daily Time Spent on Site', 'Age', 'Area Income','Daily Internet Usage', 'Male']]
X = ad_data[['Daily Time Spent on Site','Age','Daily Internet Usage']]
y = ad_data['Clicked on Ad']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

**Train and fit a logistic regression model on the training set.**

```python
logistmod = LogisticRegression()
logistmod.fit(X_train,y_train)
```

### Predictions and Evaluations

```python
prdt = logistmod.predict(X_test)
```

```python
compare = y_test.to_frame()
compare['Predicted'] = prdt
compare.sort_index(inplace=True)
```

```python
moving_sum_click = compare['Clicked on Ad'].rolling(3, center=True).sum()
moving_sum_pridt = compare['Predicted'].rolling(3, center=True).sum()
ax = moving_sum_click.plot(label='Clicked on Ad', figsize=(15,4), linewidth=4, alpha=0.5)
moving_sum_pridt.plot(ax=ax, label='Predicted', linewidth=4, alpha=0.5)
plt.legend()
```

```python
compare.plot.kde()
```

**Create a classification report for the model.**

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,prdt))
```
